```python
dataset = pd.read_csv('text_emotion.csv', encoding = 'latin-1')
```

```python
dataset.columns = ['ID', 'sentiment', 'author', 'content']
```

```python
dataset = dataset.drop(['ID', 'author'], axis=1)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
```

```python
new = []
```

```python
for i in range(len(dataset)):
    print("iterration::",i)
    temp = re.sub('[^a-zA-Z]', ' ', dataset['content'][i])
    temp = temp.lower()
    temp = temp.split()
    ps = PorterStemmer()
    temp = [ps.stem(word) for word in temp if not word in stopwords.words('english')]
    temp = ' '.join(temp)
    new.append(temp)
```

```python
new_df = pd.DataFrame({'text':new})
```

```python
new_df.to_csv('clened_data_multi_class.csv')
```
