```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle
```

```python
dataset = pd.read_csv('train.csv', header = None, encoding = 'latin-1')
```

```python
dataset.columns = ['ID', 'sentiment', 'text']
```

```python
dataset = dataset.drop(['ID'],axis = 1)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
```

```python
new = []
```

```python
for i in range(len(dataset)):
    print("iterration::",i)
    temp = re.sub('[^a-zA-Z]', ' ', dataset['text'][i])
    temp = temp.lower()
    temp = temp.split()
    ps = PorterStemmer()
    temp = [ps.stem(word) for word in temp if not word in stopwords.words('english')]
    temp = ' '.join(temp)
    new.append(temp)
```

```python
y = dataset.iloc[:,0].values
```

```python
from sklearn.cross_validation import train_test_split
X_Train, X_Test, y_Train, y_Test = train_test_split(x, y, test_size = 0.1, random_state = 0)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', gamma = 0.085)
classifier.fit(X_Train, y_Train)
```

```python
y_pred = classifier.predict(X_Test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_Test, y_pred)
```
